## RFM Modelling

```python
#Recency = Latest Date - Last Inovice Data, Frequency = count of invoice no. of transaction(s), Monetary = Sum of Total 
#Amount for each customer
import datetime as dt

#Set Latest date 2011-12-10 as last invoice date was 2011-12-09. This is to calculate the number of days from recent purchase
Latest_Date = dt.datetime(2011,12,10)

#Create RFM Modelling scores for each customer
RFMScores = Rtl_data.groupby('CustomerID').agg({'InvoiceDate': lambda x: (Latest_Date - x.max()).days, 'InvoiceNo': lambda x: len(x), 'TotalAmount': lambda x: x.sum()})

#Convert Invoice Date into type int
RFMScores['InvoiceDate'] = RFMScores['InvoiceDate'].astype(int)

#Rename column names to Recency, Frequency and Monetary
RFMScores.rename(columns={'InvoiceDate': 'Recency', 
                         'InvoiceNo': 'Frequency', 
                         'TotalAmount': 'Monetary'}, inplace=True)

RFMScores.reset_index().head()
```

```python
#Split into four segments using quantiles
quantiles = RFMScores.quantile(q=[0.25,0.5,0.75])
quantiles = quantiles.to_dict()
```

```python
quantiles
```

```python
#Functions to create R, F and M segments
def RScoring(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
    
def FnMScoring(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
#Calculate Add R, F and M segment value columns in the existing dataset to show R, F and M segment values
RFMScores['R'] = RFMScores['Recency'].apply(RScoring, args=('Recency',quantiles,))
RFMScores['F'] = RFMScores['Frequency'].apply(FnMScoring, args=('Frequency',quantiles,))
RFMScores['M'] = RFMScores['Monetary'].apply(FnMScoring, args=('Monetary',quantiles,))
RFMScores.head()
```

```python
#Calculate and Add RFMGroup value column showing combined concatenated score of RFM
RFMScores['RFMGroup'] = RFMScores.R.map(str) + RFMScores.F.map(str) + RFMScores.M.map(str)

#Calculate and Add RFMScore value column showing total sum of RFMGroup values
RFMScores['RFMScore'] = RFMScores[['R', 'F', 'M']].sum(axis = 1)
RFMScores.head()
```

```python
#Assign Loyalty Level to each customer
Quant_Level = [1,2,3,4]
Loyalty_Level = ['Platinum', 'Gold', 'Silver', 'Bronze']
Score_cuts = pd.qcut(RFMScores.RFMScore, q = 4, labels = Loyalty_Level)
Score_cuts2 = pd.qcut(RFMScores.RFMScore, q = 4, labels = Quant_Level)
RFMScores['RFM_Loyalty_Level'] = Score_cuts.values
RFMScores['RFM_Catagory_Level'] = Score_cuts2.values
RFMScores.reset_index().head()
```

## K-Means Clustering

```python
#Handle negative and zero values so as to handle infinite numbers during log transformation
def handle_neg_n_zero(num):
    if num <= 0:
        return 1
    else:
        return num
#Apply handle_neg_n_zero function to Recency and Monetary columns 
RFMScores['Recency'] = [handle_neg_n_zero(x) for x in RFMScores.Recency]
RFMScores['Monetary'] = [handle_neg_n_zero(x) for x in RFMScores.Monetary]

#Perform Log transformation to bring data into normal or near normal distribution
Log_Tfd_Data = RFMScores[['Recency', 'Frequency', 'Monetary']].apply(np.log, axis = 1).round(3)
```

```python
#Data distribution after data normalization for Recency
Recency_Plot = Log_Tfd_Data['Recency']
ax = sns.displot(Recency_Plot)
```

```python
#Data distribution after data normalization for Frequency
Frequency_Plot = Log_Tfd_Data.query('Frequency < 1000')['Frequency']
ax = sns.displot(Frequency_Plot)
```

```python
#Data distribution after data normalization for Monetary
Monetary_Plot = Log_Tfd_Data.query('Monetary < 10000')['Monetary']
ax = sns.displot(Monetary_Plot)
```

```python
from sklearn.preprocessing import StandardScaler

#Bring the data on same scale
scaleobj = StandardScaler()
Scaled_Data = scaleobj.fit_transform(Log_Tfd_Data)

#Transform it back to dataframe
Scaled_Data = pd.DataFrame(Scaled_Data, index = RFMScores.index, columns = Log_Tfd_Data.columns)
```

### Finding the optimal number of clusters

A common method for determining the optimal number of clustes is "the elbow method", and is often times done through visual inspection by seeing the point of inflection of the drop of of within some of squares errors (hence the 'elbow'). 

This can also be achieved with a bit less bias by using a package called yellowbrick to show the within sume of squares drop off, but also call in the actual value through a calable attribute. 

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
```

```python
cluster_model = KMeans()
```

```python
visualizer = KElbowVisualizer(cluster_model, k=(1,10))

cluster_visual = visualizer.fit(Scaled_Data) # Fit the data to the visualizer 
cluster_visual.show()
```

```python
# Perform K-Mean Clustering or build the K-Means clustering model
# The optimal number of clusters is 3, and will call that in with cluster_visual.eblow_value
KMean_clust = KMeans(n_clusters= cluster_visual.elbow_value_, init= 'k-means++', max_iter= 1000)
KMean_clust.fit(Scaled_Data)

#Find the clusters for the observation given in the dataset
RFMScores['Cluster'] = KMean_clust.labels_
RFMScores.head()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(7,7))

##Scatter Plot Frequency Vs Recency (will do 3D chart later)
Colors = ["red", "green", "blue"]
RFMScores['Color'] = RFMScores['Cluster'].map(lambda p: Colors[p])
ax = RFMScores.plot(    
    kind="scatter", 
    x="Recency", y="Frequency",
    figsize=(10,8),
    c = RFMScores['Color']
)
```

## Visual comparison of both techniques

### Kmeans clustering with 3 segments

This technique determined the use of 3 cluseters based on the inflection change of the **within sum of squares of stanadard errors** as the point of optimal number of groups, in this case we have 3. 

```python

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)

xs = RFMScores.Recency
ys = RFMScores.Frequency
zs = RFMScores.Monetary
clusters = RFMScores.Cluster

scatter = ax.scatter(xs, ys, zs, c=clusters,cmap='tab20b')

ax.set_title("3D plot")
ax.set_xlabel('Recency')
ax.set_ylabel('Frequency')
ax.set_zlabel('Monetary')
ax.legend(*scatter.legend_elements())
plt.show()
```

### Conclusion: Which Method Should be Used for Customer Segmentation?

Like almost everything in data analysis and data sciences, there are advantages and disadvantages for either.

Using **binning** based on an arbitrary cut off point(s) in some ways simplifies the process of segmentation. You can insights on a percentile of your market. This method allows for greater control of who is allowed in/out of your customers segments.

However using **Kmeans Clustering** has advantages as well. Number of segments can still be determined aprori, or using the the inflection point of the within sum of squares, find the optimal number of clusters and the actual segment members in the most "natural" way. But why is that beneficial?

Allowing Kmeans to determine the cluster amount and members creates groups that could hint at important, powerful insights, and prompt questions for further analysis; How are the members of each cluster alike and what characterstics do they share? And, how are the cluster members different from each other/what characterstics are set the apart? 

Understanding those questions by using unspervised machine learning lends to powerful understanding and huerstics of the market.
